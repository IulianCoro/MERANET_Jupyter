# Analysis of bruteforce data 1 element on domain VS 9 elements on domain

Input variables:
- 4x4 = 16 domains with symmetry along X and Y axis
- 50% soft material (material 1) and 50% hard material (material 2)
- E1 = 1 - elastic modulus of soft material
- E2 = 10 - elastic modulus of hard materials
- niu1 = 0.4 - Poisson's ratio of soft material
- niu2 = 0.2 - Poisson's ratio of hard material
- QUAD4 elements with full integration

```python
# import necessary python modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read data from excel files
# we have 2 files with results of brute force approach with 1 element on domain and 9 elements on domain

df_1el = pd.read_excel('BRUTE_FORCE_1EL_DOMAIN.xlsx')
df_9el = pd.read_excel('BRUTE_FORCE_9_EL_DOMAIN20220318-050338.xls')
print(df_1el.head())
print(df_9el.head())
```

From the 2 dataframes containing the data from brute force for 1 element on domain and 9 elements on domain we will create one data frame in which we will have one independent observation for each combination. 

Additionally, we check which solutions have hinges and which not. We say the a solutions has a hinge if it has 2 soft material domains or 2 hard material domains that are connected to each other by a node.

```python
# prepare and clean data

df_1el.rename(columns={'E1_eff':'E1_eff_1el', 'NIU_12': 'NIU_12_1el', 'E2_eff': 'E2_eff_1el', 'NIU_21': 'NIU_21_1el', 'G12_eff': 'G12_eff_1el', 'F_OBJ': 'F_OBJ_1el'}, inplace=True)
df_9el.rename(columns={'E1_eff':'E1_eff_9el', 'NIU_12': 'NIU_12_9el', 'E2_eff': 'E2_eff_9el', 'NIU_21': 'NIU_21_9el', 'G12_eff': 'G12_eff_9el', 'F_OBJ': 'F_OBJ_9el'}, inplace=True)
df_9el.drop(['COMB', 'EL1', 'EL2', 'EL3', 'EL4', 'EL5', 'EL6', 'EL7', 'EL8', 'EL9', 'EL10', 'EL11', 'EL12', 'EL13', 'EL14', 'EL15', 'EL16'], axis=1, inplace=True)
df = pd.concat([df_1el, df_9el], axis = 1)
# print(df.head())

# Defining the NoHinge function
def m(x, y, z):
    return abs(y-x)+abs(z-y)-abs(z-x)

def h(a, b ,c ,d):
    return m(a, b ,d) * m(a, c ,d) * m(b, a, c) * m(b, d, c)

combs = np.zeros((12870, 16))

for i in range(len(df_1el)):
    if df.EL1[i] == 2:
        combs[i, 0] = 1
    if df.EL2[i] == 2:
        combs[i, 1] = 1
    if df.EL3[i] == 2:
        combs[i, 2] = 1
    if df.EL4[i] == 2:
        combs[i, 3] = 1
    if df.EL5[i] == 2:
        combs[i, 4] = 1
    if df.EL6[i] == 2:
        combs[i, 5] = 1
    if df.EL7[i] == 2:
        combs[i, 6] = 1
    if df.EL8[i] == 2:
        combs[i, 7] = 1
    if df.EL9[i] == 2:
        combs[i, 8] = 1
    if df.EL10[i] == 2:
        combs[i, 9] = 1
    if df.EL11[i] == 2:
        combs[i, 10] = 1
    if df.EL12[i] == 2:
        combs[i, 11] = 1
    if df.EL13[i] == 2:
        combs[i, 12] = 1
    if df.EL14[i] == 2:
        combs[i, 13] = 1
    if df.EL15[i] == 2:
        combs[i, 14] = 1
    if df.EL16[i] == 2:
        combs[i, 15] = 1

def check_hinge(comb):
    sum_h = h(comb[1],comb[0],comb[5],comb[4])  + h(comb[2],comb[1],comb[6],comb[5])  + h(comb[3],comb[2],comb[7],comb[6]) + h(comb[5],comb[4],comb[9],comb[8])  + h(comb[6],comb[5],comb[10],comb[9]) + h(comb[7],comb[6],comb[11],comb[10]) +  h(comb[9],comb[8],comb[13],comb[12])+ h(comb[10],comb[9],comb[14],comb[13])+h(comb[11],comb[10],comb[15],comb[14])
    if sum_h == 0:
        return False
    else:
        return True


hinge = np.full((12870, 1), False, dtype=bool)
for i in range(12870):
    if check_hinge(combs[i]):
        hinge[i] = True
# print(hinge)

df['hinge'] = hinge
# print(df.head())
    
df.drop(['EL1', 'EL2', 'EL3', 'EL4', 'EL5', 'EL6', 'EL7', 'EL8', 'EL9', 'EL10', 'EL11', 'EL12', 'EL13', 'EL14', 'EL15', 'EL16'], axis=1, inplace=True)
print(df.head())
# print(len(df))
# df.drop_duplicates(subset=['F_OBJ'], inplace=True, ignore_index=True)
```

Using the describe method we can get some insights regarding our dataframe like mean value, standard deviation, minimum and maximum value and the interquartile range for each column.

We can see below that the mean value of the objective function drops from 2.3599 for 1 element on domain to 2.1312 for 9 elements on domain. We have seen from the mesh refinment test that solutions with hinges are more susceptible to big changes in objective function value if we increase the number of elements on domain. The origin of the checkerboard patterns is related to features of the finite element approximation, and more specifically is due to bad numerical modelling that overestimates the stiffness of checkerboards. 

We must check the difference in mean value of the objective function for 1 and 9 elements on domain for each category, with hinge and with no hinge.

```python
print(df.describe())
```

```python
mean_obj_1el_hinge = np.mean(df.F_OBJ_1el[df.hinge==True])
mean_obj_9el_hinge = np.mean(df.F_OBJ_9el[df.hinge==True])
mean_obj_1el_no_hinge = np.mean(df.F_OBJ_1el[df.hinge==False])
mean_obj_9el_no_hinge = np.mean(df.F_OBJ_9el[df.hinge==False])

print('Mean F_OBJ for 1 element on domain with hinge = ', mean_obj_1el_hinge)
print('Mean F_OBJ for 9 element on domain with hinge = ', mean_obj_9el_hinge)
RE_hinge = ((mean_obj_1el_hinge - mean_obj_9el_hinge) / mean_obj_1el_hinge) * 100
print('Relative error for solutions with hinge = ', RE_hinge)
print('Mean F_OBJ for 1 element on domain with no hinge = ', mean_obj_1el_no_hinge)
print('Mean F_OBJ for 9 element on domain with no hinge = ', mean_obj_9el_no_hinge)
RE_no_hinge = ((mean_obj_1el_no_hinge - mean_obj_9el_no_hinge) / mean_obj_1el_no_hinge) * 100
print('Relative error for solutions with no hinge = ', RE_no_hinge)
```

We can see that the relative error for solutions with hinge / checkerboard patterns is double the relative error for solutions with no hinge.
